```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix
import pylab as plt 
#Creating x and y variables.
x = df.drop(columns=['DEATH_EVENT']) 
y = df['DEATH_EVENT'] 
x = x[['age', 'anaemia', 'creatinine_phosphokinase', 'diabetes', 'ejection_fraction', 'high_blood_pressure', 'platelets', 'serum_creatinine', 'serum_sodium', 'sex', 'smoking', 'time']] 
#Splitting data into training and testing. 
x_train, x_test, y_train, y_test = train_test_split(x, y, random_state=7)
clf = KNeighborsClassifier(5) 
fit = clf.fit(x_train, y_train) 
y_pre = fit.predict(x_test) 
cm = confusion_matrix(y_test,y_pre)
print(cm)
print(clf.score(x_test,y_test)) 
from sklearn.metrics import classification_report 
print(classification_report(y_test,y_pre))
```

```python
#Hill climbing Technique for feature selection
from sklearn.utils import shuffle
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import train_test_split
x = df.drop(columns=['DEATH_EVENT']) 
y = df['DEATH_EVENT'] 
x = x[['age', 'anaemia', 'creatinine_phosphokinase', 'diabetes', 'ejection_fraction',
       'high_blood_pressure', 'platelets', 'serum_creatinine', 'serum_sodium', 'sex', 'smoking', 'time']] 
new_Ind = []
cur_MaxScore = 0.0
col_num = x.shape[1]
col_Ind_Random = shuffle(range(0,col_num), random_state=1)
for cur_f in range(0, col_num):
# here append will append to the above column 
    new_Ind.append(col_Ind_Random[cur_f])
#doing this by selecting one column and testing with all rows
    newData = x.iloc[:, new_Ind].copy()
    x_train, x_test, y_train, y_test = train_test_split(newData,y,test_size=0.2,  random_state=1)
    clf = KNeighborsClassifier(5)
    fit = clf.fit(x_train, y_train)
#.score tests the new model and compare the prediction with actual results
    cur_Score = clf.score(x_test, y_test)
    if cur_Score < cur_MaxScore:
# model accuracy didnt improve with the addition of the current feature
        new_Ind.remove(col_Ind_Random[cur_f])
    else:
# model accuracy improved
        cur_MaxScore = cur_Score
        print("Score with " + str(len(new_Ind)) + " selected features: " + str(cur_Score))
```

```python
#decision tree classification model
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix
import pylab as plt

X_train, X_test, Y_train, Y_test =  train_test_split(x, y,  test_size=0.2, random_state=2)
clf_D=DecisionTreeClassifier(random_state=2)
fit_D=clf_D.fit(X_train,Y_train)
Y_prediction=fit_D.predict(X_test)
confu_matrix=confusion_matrix(Y_test,Y_prediction)
print(confu_matrix)
print(clf_D.score(X_test,Y_test))
from sklearn.metrics import classification_report 
print(classification_report(Y_test,Y_prediction))
```

```python
#how to increase precision
from sklearn.tree import DecisionTreeClassifier
X_train,X_test, Y_train, Y_test = train_test_split(x, y, test_size=0.20, random_state=2)
clf_1 = DecisionTreeClassifier(criterion='entropy',max_depth=2,random_state=2)
fit = clf_1.fit(X_train, Y_train)
y_pre_2 = fit.predict(X_test)
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(Y_test, y_pre_2)
print(cm)
print(clf_1.score(X_test,Y_test))
from sklearn.metrics import classification_report
print(classification_report(Y_test,y_pre_2))
```
